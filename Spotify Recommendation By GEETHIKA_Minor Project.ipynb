## Creating the Model by using the KMeans Clustering

```python
from sklearn.cluster import KMeans
km=KMeans(n_clusters=5)
```

```python
#Storing the dataframe to a temporary variable
```

```python
df_temp=df
df_temp
```

```python
#Converting the String objects into numeric values
```

```python
df_temp["playlist_genre"].replace({"pop":0,"rap":1,"rock":2,"latin":3,"r&b":4,"edm":5},inplace=True)
df_temp
```

```python
#Fitting the input for the Kmeans_clustering
```

```python
km.fit(df_temp[["track_popularity","playlist_genre"]])
```

```python
km.cluster_centers_
```

```python
#Creating the clustering for Classifying the data
```

```python
df_temp["cluster"]=km.labels_
df_temp
```

```python
df_temp["cluster"].value_counts()
```

```python
#Graphical Representation of KMeans_Clustering
```

```python
sns.scatterplot(x="playlist_genre",y="track_popularity",hue="cluster",data=df_temp)
```
